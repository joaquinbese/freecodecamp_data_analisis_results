```python

# Obtener el pas con el mayor porcentaje de personas que ganan ms de 50K
country_counts = df['native-country'].value_counts()
print(country_counts)
salary_counts = df[df['salary'] == '>50K']['native-country'].value_counts()
print(salary_counts)

percentage = (salary_counts / country_counts).max()
print(percentage)

highest_earning_country = salary_counts[salary_counts / country_counts == percentage].index[0]
print(f"ss {highest_earning_country}")
# Redondear el porcentaje a 1 decimal
highest_earning_country_percentage = round(100 * percentage, 1)

# Mostrar los resultados
print("El pas con el mayor porcentaje de personas que ganan ms de 50K es:", highest_earning_country)
print("Porcentaje:", highest_earning_country_percentage, "%")
```
